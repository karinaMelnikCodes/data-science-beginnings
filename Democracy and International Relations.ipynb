```python
import numpy as np
import matplotlib.pyplot as plt; plt.rcdefaults()
import seaborn as sns
import pandas as pd

filepath = "C:/Users/19177/Documents/UChicago/Law Research/Book Project/DemocracyQuintile.xlsx"
orig = pd.read_excel(filepath)
df = pd.DataFrame(orig)
df
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np
import pandas as pd
filepath = "C:/Users/19177/Documents/UChicago/Law Research/Book Project/Chap 2/Table1.xlsx"

df = pd.read_excel(filepath, index_col=0)
dfc1 = df['% of Treaties']
dfc2 = df['Probability of concluding treaty per year']


fig = plt.figure() # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.3

dfc1.plot(kind='bar', color='#babdbd', edgecolor='black', ax=ax, width=width, position=0)
ax.set_ylim(ymin=0, ymax=70)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
dfc2.plot(kind='bar', color='#8c9596', edgecolor='black', ax=ax2, width=width, position=1)
ax2.set_ylim(ymin=0, ymax=0.40)
ax2.yaxis.set_major_formatter(mtick.PercentFormatter())

ax.set_xlim(ax.get_xlim()[0]+0.1, ax.get_xlim()[1]+0.2) 

# xticks = ax.get_xticks()
# xticklabels = ax.get_xticklabels()
ax.set_xticklabels(["Both parties democratic", "Half-democratic", "Both parties authoritarian"], rotation="horizontal")
# ax.set_xticklabels(["Both parties democratic", "Half-democratic", "Both parties authoritarian"])
# plt.xticks(xticks, xticklabels, rotation=70)

# ax.set_ylim([0, 70])
# ax2.set_ylim([0, 70])

ax.tick_params(axis='y', labelcolor='#babdbd')
ax2.tick_params(axis='y', labelcolor='#8c9596')
ax.set_ylabel('% of Treaties', color='#babdbd')
ax2.set_ylabel('Probability of concluding treaty per year', color='#8c9596')



plt.show()
```
